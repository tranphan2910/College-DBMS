```python
from faker import Faker
import random

fake = Faker()

def generate_student_id():
    """
    Generate a student ID with the format {department}{major}{school abbreviation}{year}{random number}
    """
    department_dict = {
    "IT": ["DS", "CE", "CS", "NE"], "BA": ["BA", "HM", "IB", "MK"], "BT": ["AR", "BT", "FT"],
    "CE": ["CE", "CM"], "EE": ["EE", "AC"], "EN": ["EN"], "IE": ["IE","LS"], "PH": ["SE"],
    }

    department = random.choice(list(department_dict.keys()))

    # choose a random value from the list associated with the random key
    major = random.choice(department_dict[department])

    school_abbreviation = "IU"
    year = fake.random_int(min=19, max=22)
    random_number = fake.random_int(min=100, max=2000)
    return f"{department}{major}{school_abbreviation}{year}{random_number}"

def generate_student_data(num_students):
    students = []
    for i in range(num_students):
        student_id = generate_student_id()
        major = student_id[2:4]
        department = student_id[:2]
        dob = fake.date_of_birth(minimum_age = 18, maximum_age = 21)
        # gender = fake.random_element(elements=('Male', 'Female'))
        students.append([student_id, fake.first_name() + " " + fake.last_name(), dob,fake.email(), department, major])
    return students

def generate_instructor_data(num_instructors):
    instructors = []
    for i in range(1, num_instructors+1):
        department_dict = {"IT": ["School of Computer Science and Engineering"], "BA" : ["School of Business"], "EFA" : ["School of Economics, Finance and Accounting"],
                           "EE" : ["School of Electrical Engineering"], "BT" : ["School of Biotechnology"], "IE" : ["School of Industrial Engineering and Management"], 
                           "CE" : ["School of Civil Engineering and Management"], "EN" : ["School of Languages"], "MA" : ["Department of Mathematics"],
                           "PH" : ["Department of Physics"] }
        department_id = random.choice(list(department_dict.keys()))
        department = random.choice(department_dict[department_id])

        random_position = random.choices(["Dean", "Vice Dean", "Lecture"], weights = [0.05,0.1,0.85], k = 1)
        position = random_position[0]
        dob = fake.date_of_birth(minimum_age = 27, maximum_age = 59)

        instructors.append(["INS" + str(i), fake.first_name() + " " + fake.last_name(), dob,fake.email(), department, department_id, position])
    return instructors

# def generate_course_data(num_subjects, courses):
#     subjects = []
#     for i in range(num_subjects):
#         subjects.append([i, fake.job(), random.choice([x[0] for x in courses])])
#     return subjects

def generate_subjects_data(num_courses, instructors):
    subjects = []
    for i in range(num_courses):
        number_of_students = random.choice([30, 40, 50, 60, 80, 100])
        subjects.append([random.choice([x[5] for x in instructors]) + str(i) + "IU", fake.bs(), fake.random_int(min=1, max=4), random.choice([x[1] for x in instructors]), number_of_students])
    return subjects


def generate_registration_data(num_registrations, students, subjects):
    registrations = []
    for i in range(1, num_registrations+1):
        registrations.append([i, random.choice([x[0] for x in students]), random.choice([x[0] for x in subjects]) , fake.date_this_decade()])
    return registrations


students = generate_student_data(1000)
instructors = generate_instructor_data(100)
subjects = generate_subjects_data(100, instructors)
courses = pd.read_csv("major_table.csv")
registrations = generate_registration_data(500, students, subjects)
```
